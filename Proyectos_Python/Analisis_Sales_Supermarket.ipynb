# Task 2: Univariate Analisys
Question 1 = What does the customer distribution look like?  Is it skewed?

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], kde=True)

plt.axvline(x=np.mean(df['Rating']), color='r', linestyle='--', label='Mean')
plt.axvline(x=np.percentile(df['Rating'], 25), color='g', linestyle='--', label='25th Percentile')
plt.axvline(x=np.percentile(df['Rating'], 75), color='g', linestyle='--', label='75th Percentile')

plt.title('Distribution of Ratings with Mean and Percentiles')
plt.legend()
plt.show()
```

```python
df.hist(figsize=(10,10))

```

Question 2: Do aggregate sales numbers differ much between branches?

```python
plt.figure(figsize=(8,5))
sns.countplot(x= 'Branch',hue='Branch', data=df, palette=['#1f77b4', '#ff7f0e', '#2ca02c'] )
plt.title('Number of Sales per Branch')
plt.show()
```

```python
df['Branch'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Payment',hue='Payment', data=df, palette=["#f33434","#054585","#034c03"] )
plt.title('Different Payment Methods Used')
plt.show()
```

```python
df['Payment'].value_counts()
```

# Task 3: Bivariate Analisys
Question 3: Is there a relationship between gross income and customer raitings?

```python
sns.regplot(x=df['Rating'], y=df['gross income'])

```

Question 4: Is there a noticiable time trend gross income?

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Branch',
    y='gross income',
    hue='Branch',
    data=df,
    palette=['#1f77b4', '#ff7f0e', '#2ca02c'],
    legend=False
)
plt.title("Gross Income Distribution by Branch")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Gender',
    y='gross income',
    hue='Gender',
    data=df,
    palette=["#04416d", "#9AC304"],
    legend=False
)
plt.title("Gross Income Distribution by Gender")
plt.show()
```
